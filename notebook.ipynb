**Data Cleaning & Preparation**

```python
import pandas as pd

df = pd.read_csv("IPL.csv")

# Display first five rows
df.head()
```

```python
df.isnull().sum()
```

```python
# Remove the unnecessary index column
df.drop(columns=['Unnamed: 0'], inplace=True)
text_cols = [
    'extra_type',
    'wicket_kind',
    'player_out',
    'fielders',
    'review_batter',
    'team_reviewed',
    'review_decision',
    'superover_winner',
    'result_type',
    'method',
    'new_batter',
    'next_batter'
]

#Fill missing categorical values with "Not Applicable"
for col in text_cols:
    df[col] = df[col].fillna("Not Applicable")
numeric_cols = [
    'team_runs',
    'team_balls',
    'team_wicket',
    'batter_runs',
    'batter_balls',
    'bowler_wicket'
]

for col in numeric_cols:
    df[col] = df[col].fillna(0)
#the column contains no useful information Drop it
df.drop(columns=['power_surge_start'], inplace=True)
```

```python
text_columns = [
    'match_type', 'event_name', 'batting_team', 'bowling_team',
    'batter', 'bowler', 'extra_type', 'non_striker',
    'wicket_kind', 'player_out', 'fielders',
    'review_batter', 'team_reviewed', 'review_decision',
    'player_of_match', 'match_won_by', 'win_outcome',
    'toss_winner', 'toss_decision', 'venue', 'city',
    'season', 'gender', 'team_type', 'superover_winner',
    'result_type', 'method', 'event_match_no', 'stage',
    'match_number', 'new_batter', 'next_batter', 'striker_out'
]

for col in text_columns:
    if col in df.columns:
        df[col] = df[col].fillna("Not Applicable").astype(str).str.strip().str.title()
```

```python
# Convert the date column to datetime format.
df['date'] = pd.to_datetime(df['date'])
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.rename(columns={
    'runs_batter':'batter_runs_scored',
    'runs_total':'total_runs',
    'runs_bowler':'bowler_runs'
}, inplace=True)
```

```python
 #Create a new column to classify deliveries as boundaries or non-boundaries.
 df['Boundary'] = df['batter_runs_scored'].apply(
    lambda x: 'Boundary' if x in [4,6] else 'Other'
)
```

```python
df.to_csv("cleaned_dataset.csv", index=False)
```
